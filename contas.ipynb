```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,1.5,0.001)
y = np.asin(2/3*x)*180/np.pi
```

```python
sns.set_style('whitegrid')
sns.set_context("notebook")
ax=sns.lineplot(x=x,y=y)
ax.set_ylim(0,90)
ax.set_xlim(0,1.5)
ax.set_xticks([0,0.5,1,1.5])
ax.set_yticks([0,20,40,60,90])
ax.set_ylabel(r"$\theta_c$")
ax.set_xlabel("H/h")
x1 = np.sqrt(3)/2
y1 = np.asin(2/3*x1)*180/np.pi
plt.axvline(x1,ymax=y1/90,ls='--',c='r')
plt.axhline(y1,xmax=x1/1.5,ls='--',c='r')
plt.annotate(r"$H=\frac{\sqrt{3}}{2}h$"+"\n"+r"$\theta_c = 35.3^\circ$",
             xy=(x1,y1),xytext=(0.52,65),
            arrowprops={'width':1,'shrink':0.1,'color':'r','headwidth':5})
plt.annotate("$H=h$\n"+r"$\theta_c = 41.8^\circ$",
             xy=(1,42),xytext=(1.1,65),
            arrowprops={'width':1,'shrink':0.1,'color':'r','headwidth':5})
ax.set_title("Ângulo em que o barbante perde tensão")
```
